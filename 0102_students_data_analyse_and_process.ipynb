```python
def data_process(df: pd.DataFrame):
    # 考试题目数量，当 正确数 和 错误数 非空时，相加得到每次考试题目总数
    col = '题目数'
    df[col] = 0
    df.loc[(~df['正确数'].isnull())&(~df['错误数'].isnull()),col] = df['正确数'] + df['错误数']
    
    # '类型'与'类型名称'，相同含义，保留一个即可
    category_columns = ['student_type','课程', '章节', '执照考试结果']
    numeric_columns = ['类型', '正确数', '错误数', '题目数','得分', '测试用时（分钟）']
    
    ###考试 创建时间 年月日信息提取 ###
    df['创建时间'] = pd.to_datetime(df['创建时间'])
    df['year'] = df['创建时间'].dt.year
    df['month'] = df['创建时间'].dt.month
    df['day'] = df['创建时间'].dt.day
    df['hour'] = df['创建时间'].dt.hour
    # df['rank'] = df.groupby(df['student_name'])['创建时间'].rank(ascending=False)
    
    ### student_type one hot ###
    # student_type = 滞留中 1
    # student_type != 滞留中 0
    col = 'student_type_stop'
    df[col] = 0
    df.loc[df['student_type']=='滞留中', col] = 1    

    # 考过生
    col = 'student_type_pass'
    df[col] = 0
    df.loc[df['student_type']=='考过生', col] = 1

    # 在学生
    col = 'student_type_learning'
    df[col] = 0
    df.loc[df['student_type']=='在学生', col] = 1

    # 考试学员
    col = 'student_type_exam'
    df[col] = 0
    df.loc[df['student_type']=='考试学员', col] = 1
    
    ### 日常章节考试 是否通过 one hot###
    # 是否通过 = 通过 1
    # 是否通过 = 未通过 0
    col = 'pass_label'
    df[col] = 0
    df.loc[df['是否通过']=='通过', col] = 1
    df.loc[df['是否通过']=='未通过', col] = 0
    
    # 1）解剖与机能
    # 2）评估与治疗
    # 3）手法与实践
    # 4）病理与禁忌
    # 5）专业与道德
    # 6）新增测试题
    # 7）测前评估
    
    # [课程顺序，章节数目]
    course_order = {
        '解剖与机能': [1, 16],
        '评估与治疗': [2, 14],
        '手法与实践': [3, 15],
        '病理与禁忌': [4, 16],
        '专业与道德': [5, 16],
        '新增测试题': [6, None],
        '测前评估': [7, None],
    }
    
    col1 = 'course_order' # 课程顺序
    col2 = 'chapter_number' # 章节数目
    df[col1] = 0
    df[col2] = 0
    for k in course_order:
        df.loc[(~df['课程'].isnull())&(df['课程'].str.contains(k)), col1] = course_order[k][0]
        df.loc[(~df['课程'].isnull())&(df['课程'].str.contains(k)), col2] = course_order[k][1]
    
    
    chapter_order = {'Chapter '+str(i): i for i in range(1,20)}
    
    col = 'chapter_order' # 章节顺序
    for k in chapter_order:
        df.loc[(~df['章节'].isnull())&(df['章节'].str.contains(k)), col] = chapter_order[k]
    df.loc[(~df['章节'].isnull())&(df['章节'].str.contains('(Final)')), col] = df['chapter_number']
    
    col = 'is_final_exam'
    df[col] = 0
    df.loc[(~df['章节'].isnull())&(df['章节'].str.contains('(Final)')), col] = 1
    df.loc[(~df['课程'].isnull())&(df['课程'].str.contains('新增测试题|测前评估')), col] = 1
    
    # 中文字段 转 英文名
    rename_dict = {
        '课程': 'course',
        '章节': 'chapter',
        '类型': 'exam_type',
        '类型名称': 'exam_type_name',
        '正确数': 'correct_number',
        '错误数': 'error_number',
        '得分': 'score',
        '测试用时（分钟）': 'exam_time_minutes',
        '是否通过': 'pass',
        '创建时间': 'exam_date',
        '备注': 'notes',
        '执照考试时间': 'license_exam_date',
        '执照考试结果': 'license_exam_pass',
        '题目数': 'items_number',
        }
    df.rename(columns=rename_dict, inplace=True)
    
    # dataframe 字段顺序调整
    cols = [
        # 学生信息
        'file_name',
        'student_name',
        'student_type',
        
        # 课程章节信息
        'course',
        'course_order',
        'chapter_number',
        'chapter',
        'chapter_order',
        'is_final_exam',
        
        # label信息，考试是否通过，学生类型
        'pass',
        'student_type_stop',
        'student_type_pass',
        'student_type_learning',
        'student_type_exam',
        'pass_label',
        
        # 考试信息
        'exam_type',
        'exam_type_name',
        'items_number',
        'correct_number',
        'error_number',
        'score',
        'exam_time_minutes',
        'exam_date',
        'year',
        'month',
        'day',
        'hour',
        'license_exam_date',
        'license_exam_pass',
        'notes',
    ]
    
    df = df[cols]
    
    # df = df.fillna(0)
    return df
```

```python
data_static = spark.sql("""
select
file_name,
student_name,
student_type,
course,
course_order,
chapter_number,
chapter,
chapter_order,
is_final_exam,
pass,
student_type_stop,
student_type_pass,
student_type_learning,
student_type_exam,
pass_label,
exam_type,
exam_type_name,
items_number,
correct_number,
error_number,
score,
exam_time_minutes,
exam_date,
year,
month,
day,
hour,
license_exam_date,
license_exam_pass,
notes,

-- score
round(sum(score) over(w_last_5_times), 2) as last_5_score_sum,
round(avg(score) over(w_last_5_times), 2) as last_5_score_avg,
round(max(score) over(w_last_5_times), 2) as last_5_score_max,
round(min(score) over(w_last_5_times), 2) as last_5_score_min,
round(sum(score) over(w_history), 2) as history_score_sum,
round(avg(score) over(w_history), 2) as history_score_avg,
round(max(score) over(w_history), 2) as history_score_max,
round(min(score) over(w_history), 2) as history_score_min,
count(1) over(w_history) as exam_times,

-- correct_number
round(sum(correct_number) over(w_last_5_times), 2) as last_5_correct_number_sum,
round(avg(correct_number) over(w_last_5_times), 2) as last_5_correct_number_avg,
round(max(correct_number) over(w_last_5_times), 2) as last_5_correct_number_max,
round(min(correct_number) over(w_last_5_times), 2) as last_5_correct_number_min,
round(sum(correct_number) over(w_history), 2) as history_correct_number_sum,
round(avg(correct_number) over(w_history), 2) as history_correct_number_avg,
round(max(correct_number) over(w_history), 2) as history_correct_number_max,
round(min(correct_number) over(w_history), 2) as history_correct_number_min,

-- error_number
round(sum(error_number) over(w_last_5_times), 2) as last_5_error_number_sum,
round(avg(error_number) over(w_last_5_times), 2) as last_5_error_number_avg,
round(max(error_number) over(w_last_5_times), 2) as last_5_error_number_max,
round(min(error_number) over(w_last_5_times), 2) as last_5_error_number_min,
round(sum(error_number) over(w_history), 2) as history_error_number_sum,
round(avg(error_number) over(w_history), 2) as history_error_number_avg,
round(max(error_number) over(w_history), 2) as history_error_number_max,
round(min(error_number) over(w_history), 2) as history_error_number_min,

-- exam_time_minutes
round(sum(exam_time_minutes) over(w_last_5_times), 2) as last_5_exam_time_minutes_sum,
round(avg(exam_time_minutes) over(w_last_5_times), 2) as last_5_exam_time_minutes_avg,
round(max(exam_time_minutes) over(w_last_5_times), 2) as last_5_exam_time_minutes_max,
round(min(exam_time_minutes) over(w_last_5_times), 2) as last_5_exam_time_minutes_min,
--
round(sum(exam_time_minutes) over(w_history), 2) as history_exam_time_minutes_sum,
round(avg(exam_time_minutes) over(w_history), 2) as history_exam_time_minutes_avg,
round(max(exam_time_minutes) over(w_history), 2) as history_exam_time_minutes_max,
round(min(exam_time_minutes) over(w_history), 2) as history_exam_time_minutes_min,

sum(CASE WHEN pass = '通过' THEN 1 ELSE 0 END) over(w_last_5_times) as last_5_exam_pass_cnt,
sum(CASE WHEN pass = '未通过' THEN 1 ELSE 0 END) over(w_last_5_times) as last_5_notexam_pass_cnt,
sum(CASE WHEN pass = '通过' THEN 1 ELSE 0 END) over(w_history) as history_pass_cnt,
sum(CASE WHEN pass = '未通过' THEN 1 ELSE 0 END) over(w_history) as history_notpass_cnt,

-- Example: Success rate for the last 5 exams
round(CASE WHEN count(1) over(w_last_5_times) > 0 THEN
  sum(CASE WHEN pass = '通过' THEN 1 ELSE 0 END) over(w_last_5_times) / nullif(count(1) over(w_last_5_times), 0)
ELSE NULL END, 2) as last_5_exam_success_rate,
-- Example: Success rate for the entire history
round(sum(if(pass='通过',1,0)) over(w_history) / count(1) over(w_history), 2) as history_exam_success_rate,
-- Example: Score improvement over the last 5 exams
round(score - lag(score, 5, 0) over(partition by student_name order by exam_date), 2) as score_improvement_over_last_5_exams,
-- Example: Increase in error number over history
round(lag(error_number, 1, 0) over(partition by student_name order by exam_date) - error_number, 2) as error_number_increase_over_history,
-- Total exam time for the last 5 exams
sum(exam_time_minutes) over(w_last_5_times) as last_5_total_exam_time,
-- Time since last exam
datediff(exam_date, lag(exam_date, 1) over(partition by student_name order by exam_date)) as days_since_last_exam,
-- Total exams attempted by student in the course
count(exam_date) over(partition by student_name, course) as total_exams_course

from table
-- where student_name in ('jiang yu', 'Jinhui Yang')
window w_last_5_times as (partition by student_name order by exam_date rows between 5 preceding and 1 preceding),
       w_history as (partition by student_name order by exam_date rows unbounded preceding)
order by student_name, exam_date

""")
data_static.show(5)
```
